Our Final Dataset consists of 2 datasets that we have joined and prepared for machine learning purposes. The first can be found on Kaggle at this link:
https://www.kaggle.com/datasets/buntyshah/auto-insurance-claims-data

This data was joined along the 'State' column with population data from the US Census. This official data can be found at this link:
https://www.census.gov/programs-surveys/geography/guidance/geo-areas/urban-rural.html

Our goal is to use this data to predict whether or not an insurance claim would result in insurance fraud being reported.

```python
#import modules
import numpy as np
import pandas as pd

#read dataset into environment
final_dataset = pd.read_csv("final_dataset.csv")
```

The prepared data can be found in the file 'final_dataset.csv'. The target column 'fraud_reported' initially contains character values 'Y' and 'N' to represent positive or negative outcomes. For our purposes, we decided to convert these to 1 and 0 integer values.

```python
#convert Y and N to 1 and 0
for i in range(0, len(final_dataset["fraud_reported"])):
    if final_dataset["fraud_reported"][i] == 'Y':
        final_dataset["fraud_reported"][i] = 1
    elif final_dataset["fraud_reported"][i] == 'N':
        final_dataset["fraud_reported"][i] = 0
    else:
        pass
```

This is the dataset at this point in our preprocessing.

```python
#show dataframe
final_dataset
```

We have several columns that are of non-numeric datatypes. These will have to be handled separately from our numeric data and most will require encoding.

```python
#show dataframe info
final_dataset.info()
```

```python
#rename problemnatic column names
final_dataset = final_dataset.rename({'2020 \r\nURBAN POP': '2020 URBAN POP', '2020 \r\nRURAL POP': '2020 RURAL POP', '2010 \r\nTOTAL POP': '2010 TOTAL POP', '2010 \r\nURBAN POP': '2010 URBAN POP', '2010\r\n RURAL POP': '2010 RURAL POP'}, axis=1)
```

An irregularity in the naming convention was spotted in the data gathered from the US Census dataset. Some column names were found to contain carriage return and line feed characters. This was corrected by renaming the columns.

```python
#show dataset info
final_dataset.info()
```

Columns containing dates were converted to numeric values by removing the hyphens and converting to an 'int' data type. This was done as to retain the inferred hierarchy between dates (i.e. 20200420 > 20200419). We did not want to lose that information through ordinal encoding.

```python
#convert dates to integers
def removeHyphens(string):
    string = string.replace("-", "")
    return int(string)

for i in range(len(final_dataset)):
    final_dataset["policy_bind_date"][i] = removeHyphens(final_dataset["policy_bind_date"][i])
```

```python
#convert dates to integers
for i in range(len(final_dataset)):
    final_dataset["incident_date"][i] = removeHyphens(final_dataset["incident_date"][i])
```

```python
#show empty column
final_dataset['_c39']
```

The column '_c39' was completely empty. This column was dropped as it served no purpose.

```python
#drop empty column
final_dataset.drop("_c39", axis=1, inplace=True)
```

```python
#show column that will be changed
final_dataset["policy_csl"]
```

The 'policy_csl' column was made up of strings that represented fractions. These strings were converted to numeric values that still maintained their intended values.

```python
#change fractoins to decimals
def cslHandle(fraction):
    fraction = int(fraction.split('/')[0]) / int(fraction.split('/')[1])
    return fraction

for i in range(len(final_dataset)):
    final_dataset["policy_csl"][i] = cslHandle(final_dataset["policy_csl"][i])

#show changed column
final_dataset["policy_csl"]
```

```python
#show locaiton column
final_dataset['incident_location']
```

```python
#drop location column
final_dataset.drop("incident_location", axis=1, inplace=True)
```

Incident location contained strings that represented the location of the incident. This was deemed too internally inconsistent to be of any real use in machine learning and was dropped.

```python
#show dataset info at this stage of preprocessing
final_dataset.info()
```

We categorized out columns based on how we will be preprocessing their data. This is also where we split our data.

```python
#import train test split
from sklearn.model_selection import train_test_split

#split
train_set, test_set = train_test_split(final_dataset, test_size=0.2, random_state=42)


#categorize attributes
num_std_attribs = ['months_as_customer', 'age', 'policy_number', 'policy_bind_date', 'policy_csl', 'policy_deductable', 'policy_annual_premium', 'umbrella_limit', 'capital-gains', 'capital-loss', 'incident_date', 'incident_hour_of_the_day', 'number_of_vehicles_involved', 'bodily_injuries', 'witnesses', 'total_claim_amount', 'injury_claim', 'property_claim', 'vehicle_claim', 'auto_year', 'STATEFP', '2020 TOTAL POP', '2020 URBAN POP', '2020 PCT URBAN POP', '2020 RURAL POP', '2020 PCT RURAL POP', '2010 TOTAL POP', '2010 URBAN POP', '2010 PCT URBAN POP', '2010 RURAL POP', '2010 PCT RURAL POP']
cat_attribs = ["insured_sex", "insured_education_level", "insured_relationship", "incident_type", "collision_type", "incident_severity", "authorities_contacted", "property_damage"]
other_attribs = ["insured_occupation", "insured_hobbies", "incident_state", "incident_city", "police_report_available", "auto_model", "auto_make", "STATE NAME", 'policy_state']

#separate labels
train = train_set.drop("fraud_reported", axis=1)
train_labels = train_set["fraud_reported"].copy()
train_labels = train_labels.astype('int')

test = test_set.drop("fraud_reported", axis=1)
test_labels = test_set["fraud_reported"].copy()
test_labels = test_labels.astype('int')
```

Numerical features were run through a median imputer. Our data contained no missing values but best practice was followed.
Categorical features were broken up through one hot encoding.
Text data that did not represent categories were encoded by ordinal encoding.

```python
#import modules to build pipelines
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder

#Numerical Features
numeric_std_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')),
                                  ('stdscaler', StandardScaler())])

#Other Features
other_pipeline = Pipeline([('ordinal', OrdinalEncoder())])

#Categorical Features
cat_pipeline = Pipeline([('onehot', OneHotEncoder())])
```

We bring our data back together in a full transformer

```python
#import modules for transformer
from sklearn.feature_selection import SelectFromModel
from sklearn.compose import ColumnTransformer

#make full transformer
full_transformer = ColumnTransformer([('numeric_stdpreprocessing', numeric_std_pipeline, num_std_attribs),
                                ('numeric_minmaxpreprocessing', other_pipeline, other_attribs),
                                 ('cat_preprocessing', cat_pipeline, cat_attribs)
                                ])
```

The 'l2_full_pipeline' utilizes the default penalty value fr LogisticRegression() which is l2.

```python
#Logistic Regression with l2 penalty
from sklearn.linear_model import LogisticRegression

l2_full_pipeline = Pipeline([('all_column_transformation', full_transformer),
                             ('feature_selection', SelectFromModel(LogisticRegression(), threshold='mean')),
                        ('log_regression', LogisticRegression())
                      ])
```

```python
#function to train and evaluate models
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve
from sklearn.metrics import confusion_matrix, auc
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
#calculate metrics to evaluate models
def fit_and_print(p, train_set, train_labels, test_set, test_labels):
  p.fit(train_set, train_labels)
  train_preds = p.predict(train_set)
  test_preds = p.predict(test_set)
  tn, fp, fn, tp = confusion_matrix(train_labels, train_preds).ravel()
  training_set_specificity = tn / (tn+fp)
  tn, fp, fn, tp = confusion_matrix(test_labels, test_preds).ravel()
  test_set_specificity = tn / (tn+fp)

  print("The training set accuracy is " + str(accuracy_score(train_preds, train_labels)))
  print("The training set precision is " + str(precision_score(train_preds, train_labels)))
  print("The training set recall is " + str(recall_score(train_preds, train_labels, zero_division = 1)))
  print("The training set f1 score is " + str(f1_score(train_preds, train_labels)))
  print("The training set specificity is " + str(training_set_specificity))

  print("The test set accuracy is " + str(accuracy_score(test_preds, test_labels)))
  print("The test set precision is " + str(precision_score(test_preds, test_labels)))
  print("The test set recall is " + str(recall_score(test_preds, test_labels, zero_division = 1)))
  print("The test set f1 score is " + str(f1_score(test_preds, test_labels)))
  print("The test set specificity is " + str(test_set_specificity))
```

```python

fit_and_print(l2_full_pipeline, train, train_labels, test, test_labels)
```

```python
none_full_pipeline = Pipeline([('all_column_transformation', full_transformer),
                              ('feature_selection', SelectFromModel(LogisticRegression(), threshold='mean', max_features=10)),
                        ('log_regression', LogisticRegression(penalty='none'))
                      ])
```

'none_full_pipeline' utilizes no penalty for LogisticRegression()

```python
fit_and_print(none_full_pipeline, train, train_labels, test, test_labels)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_full_pipeline = Pipeline([('all_column_transformation', full_transformer),
                               ('feature_selection', SelectFromModel(DecisionTreeClassifier(), threshold='mean', max_features=10)),
                        ('tree_classify', DecisionTreeClassifier())
                      ])
```

```python
fit_and_print(tree_full_pipeline, train, train_labels, test, test_labels)
```

tree_full_pipeline uses DecisionTreeClassifier(). This was prone to overfitting.

rnd_full_pipeline utilizes RandomForestClassifier()

```python
from sklearn.ensemble import RandomForestClassifier
rnd_full_pipeline = Pipeline([('all_column_transformation', full_transformer),
                              ('pca', PCA(n_components=2)),
                              ('feature_selection', SelectFromModel(RandomForestClassifier(), threshold='mean')),
                        ('log_regression', RandomForestClassifier(max_depth=7))
                      ])
```

```python
fit_and_print(rnd_full_pipeline, train, train_labels, test, test_labels)
```
